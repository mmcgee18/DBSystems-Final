Notebook to clean recipe data, results in recipe_clean.csv

```python
import pandas as pd

# Define the file path (use r'' for Windows local paths to avoid errors)
file_path = 'recipes.csv'

# Import the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first 5 rows to verify the import
df.head()
```

```python
pd.set_option('display.max_colwidth', None)
df['url']
```

```python
pd.reset_option('all')
df['directions']
```

```python
# Get base data for recipe table
recipe_df = df[['recipe_name', 'prep_time', 'cook_time', 'total_time', 'servings', 'yield', 'rating', 'url', 'img_src']]
recipe_df
```

```python
pd.set_option('display.max_colwidth', None)
df['nutrition']

# Want to extract nutrition data into new columns
# ['total_fat', 'sat_fat', 'cholesterol', 
#  'sodium', 'total_carbohydrate', 'dietary_fiber', 
#  'total_sugars', 'protein', 'vitamin_c', 'calcium', 
#  'iron', 'potassium']
```

```python
# Define the mapping of column names and their specific regex patterns
patterns = {
    'total_fat (g)': r'Total Fat (\d+g)',
    'sat_fat (g)': r'Saturated Fat (\d+g)',
    'cholesterol (mg)': r'Cholesterol (\d+mg)',
    'sodium (mg)': r'Sodium (\d+mg)',
    'total_carbohydrate (g)': r'Total Carbohydrate (\d+g)',
    'dietary_fiber (g)': r'Dietary Fiber (\d+g)',
    'total_sugars (g)': r'Total Sugars (\d+g)',
    'protein (g)': r'Protein (\d+g)',
    'vitamin_c (mg)': r'Vitamin C (\d+mg)',
    'calcium (mg)': r'Calcium (\d+mg)',
    'iron (mg)': r'Iron (\d+mg)',
    'potassium (mg)': r'Potassium (\d+mg)'
}

# Create the new nutrition_df by extracting each pattern from the original column
nutrition_df = pd.DataFrame()

for col, regex in patterns.items():
    nutrition_df[col] = df['nutrition'].str.extract(regex)

# Display the new DataFrame
nutrition_df.head()
```

```python
# Strip 'g' and 'mg' from all values in the new DataFrame
# We use regex=True to replace both units at once
nutrition_df = nutrition_df.replace(r'g|mg', '', regex=True)

# Convert the entire DataFrame to numeric (floats) for analysis
nutrition_df = nutrition_df.apply(pd.to_numeric, errors='coerce')

nutrition_df.head()
```

```python
# Append the cleaned nutrition_df columns back to the original df
recipe_df = pd.concat([recipe_df, nutrition_df], axis=1)

# Verify the result
recipe_df.head()
```

```python
recipe_df.to_csv('recipe_clean.csv', index=False)
```
